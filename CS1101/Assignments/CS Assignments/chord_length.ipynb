```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p=[]
n=int(input("Enter the number of runs: "))
rng=range(1000,n,10**4)
num_lst=list(rng)
for i in rng:
    th_1=np.random.uniform(low=0.0,high=2*np.pi,size=i)
    th_2=np.random.uniform(low=0.0,high=2*np.pi,size=i)
    c=0
    for j in range(i):
        d=((np.cos(th_1[j])-np.cos(th_2[j]))**2+(np.sin(th_2[j])+np.sin(th_1[j]))**2)**0.5
        l=3**0.3
        if d>l:
            c+=1        
    p.append(c/i)
```

```python
fig=plt.figure(figsize=(5,5),facecolor='white')
plt.plot(num_lst,p,'r',linestyle='--', marker='o')
plt.grid()
plt.xlabel('Number of runs',fontsize=14)
plt.ylabel('Probability',fontsize=14)
plt.legend(['Monte-Carlo'])
plt.show()
```
